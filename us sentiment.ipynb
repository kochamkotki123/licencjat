```python
import nltk

nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
vds=SentimentIntensityAnalyzer()
```

```python
#wgrywanko danych
import pandas as pd

df = pd.read_csv('us_merged.csv')
```

```python
df
```

```python
reviews = df['review']
```

sentyment

```python
def compound_score(review):
    scores = vds.polarity_scores(review)
    return scores['compound']
```

```python
df['compound'] = df['review'].apply(compound_score)
```

```python
df
```

zmiana liczb (compound score) na labele

```python
def get_sentiment_label(compound_score):
    if compound_score >= 0.2:
        return 'positive'
    elif compound_score <= -0.2:
        return 'negative'
    else:
        return 'neutral'
```

```python
review_sentiments = []
for review in reviews:
    scores = vds.polarity_scores(review)
    compound_score = scores['compound']
    sentiment_label = get_sentiment_label(compound_score)
    review_sentiments.append((review, compound_score, sentiment_label))
```

```python
for review, compound_score, sentiment_label in review_sentiments:
    print(f"Review: '{review}'\nCompound Score: {compound_score}\nSentiment: {sentiment_label}\n")
```

```python
#dodanie do df
df['sentyment'] = df['compound'].apply(get_sentiment_label)
```

```python
df
```

```python
#ZLICZENIE
sentiment_counts = {'positive': 0, 'negative': 0, 'neutral': 0}

for review in reviews:
    scores = vds.polarity_scores(review)
    compound_score = scores['compound']
    sentiment_label = get_sentiment_label(compound_score)
    sentiment_counts[sentiment_label] += 1

print("Podsumowanie:")
print(f"Positive: {sentiment_counts['positive']}")
print(f"Neutral: {sentiment_counts['neutral']}")
print(f"Negative: {sentiment_counts['negative']}")
```

korelacja oceny użytkowników i przypisanego sentymentu

```python
#zmiana ocen na stringi
df['rating'] = df['rating'].str.replace('/10', '').astype(int)
```

```python
#zmiana labeli na stringi
def senti_number(x):
    if x == 'positive':
        return 1
    elif x == 'neutral':
        return 0
    else:
        return -1

df['sentymenttttt'] = df['sentyment'].apply(senti_number)
```

```python
df
```

```python
correlation1 = df['compound'].corr(df['rating'])
print(f"\nKorelacja compound score i ocena: {correlation1}")
```

```python
correlation2 = df['sentymenttttt'].corr(df['rating'])
print(f"\nKorelacja label i ocena: {correlation2}")
```
