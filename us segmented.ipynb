```python
#SPRAWDZENIE JAKIE WARTOSCI MAJA SLOWA W ORYGINALNYM SLOWNIKU

vader_lexicon = vds.lexicon #wgranie slownika

def get_sentiment_score(word):
    return vader_lexicon.get(word, "Word not found in VADER lexicon")


words_to_check = ["afraid", "frightened", "alarmed", "shocked", "threatened", "horrendous", "scary", "terrified","disgusted","hair-rising","terror", "horror", "horrified",
                  "terror","distress","helpless","anxious","worried","apprehension","creepy","uneasy","fear","petrified","disturbing","chilling","eerie","intimidating","overwhelming","unnerving",
                  "bloody","gory","gore","repulsed","gross","appalling"]
```

```python
for word in words_to_check:
    print(f"Sentiment score for '{word}': {get_sentiment_score(word)}")
```

właściwa analiza

```python
#ROZSEGMENTOWANIE RECENZJI

nltk.download('punkt')

#segment text into sentences and analyze each sentence
def make_segments(review):
    sentences = nltk.sent_tokenize(review)
    # Analyze each sentence
    segment_scores = [vds.polarity_scores(sentence)['compound'] for sentence in sentences]
    #get review's average compound score
    mean_compound_score = sum(segment_scores) / len(segment_scores) if segment_scores else 0
    return mean_compound_score
```

```python
df['compound'] = df['review'].apply(make_segments)
```

```python
df
```

zmiana liczb (compound score) na labele

```python
def get_sentiment_label(compound):
    if compound >= 0.2:
        return 'positive'
    elif compound <= -0.2:
        return 'negative'
    else:
        return 'neutral'
```

```python
#dodanie do df
df['sentyment'] = df['compound'].apply(get_sentiment_label)
```

```python
df
```

```python
#ZLICZENIE
sentiment_counts = {'positive': 0, 'negative': 0, 'neutral': 0}

sentyment = df['sentyment']
for x in sentyment:
  if x=='positive':
    sentiment_counts['positive'] += 1
  elif x=='negative':
    sentiment_counts['negative'] += 1
  else:
    sentiment_counts['neutral'] += 1

print("Podsumowanie:")
print(f"Positive: {sentiment_counts['positive']}")
print(f"Neutral: {sentiment_counts['neutral']}")
print(f"Negative: {sentiment_counts['negative']}")
```

```python
#zmiana ocen na stringi
df['rating'] = df['rating'].str.replace('/10', '').astype(int)
```
